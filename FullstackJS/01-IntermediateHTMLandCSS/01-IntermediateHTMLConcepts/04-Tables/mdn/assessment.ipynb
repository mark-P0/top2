````python
_ = [
    [
        *i[:9],          # Unpack the first 9 columns
        ' '.join(i[9:])  # Join back the remaining split items as they're actually a single column
    ]
    for info in planets
    if (i := info.split(' '))  # Store split result into a variable; filtering actually (should) does nothing
]


_
````

````python
# Transpose / Rotate array
# Set columns as rows, and vice versa
_ = list(zip(*_))


_
````

````python
# Wrap text in HTML elements accordingly
_ = [
    [f'<th scope="row">{name}</th>' for name in _[0]],         # Planet names are headers
    *[
        [f'<td>{data}</td>' for data in col] for col in _[1:]  # Everything else is a regular cell
    ],
]


_
````

````python
# Pad cells according to the widest cell on the column
# Has no visual effect; mostly for crosswise layout in the HTML source
# Transforms each item in a column array into their string lengths,
# gets the maximum among those, then uses it as basis for how much
# should the f-string syntax pad the actual strings.
_ = [
    [
        f'{item:<{max(map(len, col))}}'
        for item in col
    ]
    for col in _
]


_
````

````python
# Revert array to actual tabular structure
_ = list(zip(*_))


_
````

````python
# Wraps all rows in an actual row element
_ = [f"<tr>  {' '.join(row)}  </tr>" for row in _]


_
````

````python
# Joins the rows into a single string
_ = '\n'.join(_)


print(_)
_
````
